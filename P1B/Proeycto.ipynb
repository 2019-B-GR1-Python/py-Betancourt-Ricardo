```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = ".\\DATA\\crimenydelito.csv"
df = pd.read_csv(path)
```

```python
path_guardado = ".\\DATA\\crimenydelito.pickle"
df.to_pickle(path_guardado)
df_pickle_guardado = pd.read_pickle(path_guardado)
df_pickle_guardado
```

```python

delito = np.unique(df_pickle_guardado['tipo_delito'].get_values(),return_counts=True)[0]
delito_num = np.unique(df_pickle_guardado['tipo_delito'].get_values(),return_counts=True)[1]

posicion_y = np.arange(len(delito))

plt.figure(figsize=(15,25))
plt.barh(posicion_y, delito_num, align = "center")
plt.yticks(posicion_y, delito)
plt.xlabel('Ocurrencia', fontsize=15)
plt.title("Ocurrencia por Tipo de Delito", fontsize=20)
for i, v in enumerate(delito_num):
    plt.text(v + 3, i, str(v),fontsize='x-large')
plt.show()
```

```python
armados =  np.unique(df_pickle_guardado['uso_arma'].get_values(),return_counts=True)[0]
ocurrencia = np.unique(df_pickle_guardado['uso_arma'].get_values(),return_counts=True)[1]
plt.figure(figsize=(8,8))
plt.pie(ocurrencia, labels=armados, autopct='%1.2f%%', textprops={'fontsize': 14})
plt.title("% de delitos con uso de Armas", fontsize=20)
plt.show()
```

```python
def numero_delitos_anio(array):
    aux_arreglo_anios=[0,0,0]
    aux = 0
    for i in array[0].astype('<U4'):
        if i == '2015':
            aux_arreglo_anios[0] = aux_arreglo_anios[0] +array[1][aux]
        elif i == '2016':
            aux_arreglo_anios[1] = aux_arreglo_anios[1] +array[1][aux]
        elif i == '2017':
            aux_arreglo_anios[2] = aux_arreglo_anios[2] +array[1][aux]
        aux+=1
    return aux_arreglo_anios
    
    


```

```python
fechas_ocurridos=np.unique(df_pickle_guardado['fecha'].get_values(),return_counts=True)
cantidad_crimenes_por_anio = numero_delitos_anio(fechas_ocurridos)

anios = ("2015","2016","2017")
plt.figure(figsize=(10,10))
plt.title("Cantidad de Delitos ocurridos por año", fontsize=20)
plt.xlabel("Años", fontsize=15)
plt.ylabel("Cantidad de Delitos ocurridos", fontsize=15)
indice = np.arange(len(cantidad_crimenes_por_anio))
plt.xticks(indice, anios)
plt.plot(cantidad_crimenes_por_anio)
for x,y in zip(indice,cantidad_crimenes_por_anio):
    plt.text(x,y,y,fontsize='x-large')
plt.show()
```
